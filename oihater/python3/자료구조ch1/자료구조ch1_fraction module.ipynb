```python
from fractions import Fraction
```

```python
def rounding_floats(num1, num2):
    return round(num1, num2)
```

```python
rounding_floats(126543.141592,-3)
```

```python
def float_to_fractions(num):
    return Fraction(*num.as_integer_ratio())
```

```python
float_to_fractions(2.75)
```

```python
def get_denominator(num1,num2):
    """분모반환"""
    a = Fraction(num1,num2)
    return a.denominator
```

```python
get_denominator(3,4)
```

```python
def get_numerator(num1,num2):
    a = Fraction(num1,num2)
    return a.numerator
```

```python
get_numerator(3,4)
```
